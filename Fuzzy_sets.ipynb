```python
import numpy as np  # Correct the typo here
import matplotlib.pyplot as plt

#define fuzzy sets using membership functions
x = np.arange(0, 10, 0.1)  # Also fix typo here: arange, not arrange
A = np.maximum(0, np.minimum(1, (x-2)/(6-2)))
B = np.maximum(0, np.minimum(1, (8-x)/(8-2)))

# plot fuzzy sets A and B
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='set A')
plt.plot(x, B, label='set B')
plt.xlabel('universe of discourse')
plt.ylabel('membership Degree')
plt.title('Fuzzy sets A and B')
plt.legend()
plt.show()
```

```python
union =np.maximum(A,B)
#plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='set A') # Changed 'lable' to 'label'
plt.plot(x, B, label='set B') # Changed 'lable' to 'label'
plt.plot(x, union, label='union A U B',linestyle="dashed") # Changed 'lable' to 'label'
plt.xlabel('universe of discourse')
plt.ylabel('membership degree')
plt.title('Fuzzy set union')
plt.legend()
plt.show()
```

```python
# prompt: intersection of fuzzy sets A and B

import matplotlib.pyplot as plt
import numpy as np
intersection = np.minimum(A, B)

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='set A')
plt.plot(x, B, label='set B')
plt.plot(x, intersection, label='intersection A ∩ B', linestyle="dashed")
plt.xlabel('universe of discourse')
plt.ylabel('membership degree')
plt.title('Fuzzy set intersection')
plt.legend()
plt.show()
```

```python
# prompt: compliment of fuzzy set A

import matplotlib.pyplot as plt
# Calculate the complement of fuzzy set A
complement_A = 1 - A

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='set A')
plt.plot(x, complement_A, label='complement of A', linestyle="dashed")
plt.xlabel('universe of discourse')
plt.ylabel('membership degree')
plt.title('Fuzzy set complement')
plt.legend()
plt.show()
```

```python
# prompt: fuzzy sets  for temperature levels

import matplotlib.pyplot as plt
import numpy as np
# Define the universe of discourse for temperature
temp_range = np.arange(0, 50, 0.1)

# Define fuzzy sets for temperature levels
cold = np.maximum(0, np.minimum(1, (15 - temp_range) / 15))
comfortable = np.maximum(0, np.minimum(1, np.minimum((temp_range - 15) / 10, (30 - temp_range) / 10)))
hot = np.maximum(0, np.minimum(1, (temp_range - 20) / 15))

# Plot the fuzzy sets
plt.figure(figsize=(12, 6))
plt.plot(temp_range, cold, label='Cold')
plt.plot(temp_range, comfortable, label='Comfortable')
plt.plot(temp_range, hot, label='Hot')
plt.xlabel('Temperature (°C)')
plt.ylabel('Membership Degree')
plt.title('Fuzzy Sets for Temperature Levels')
plt.legend()
plt.grld(True)
plt.show()
```
